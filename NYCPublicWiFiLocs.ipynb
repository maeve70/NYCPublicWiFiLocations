If you cut the budgets for libraries you're taking away a community hub & resource for people. Libraries have computer workstations for people to use.

Accorinding to data from 2021, and all this data here:
https://data.cityofnewyork.us/dataset/Broadband-Data-Dig-Datasets/ft4n-yqee

Public libraries are communal centers for many NYC neighborhoods, offering a wide range of programs including computer usage. If you cut the budget to the libraries, you're hurting communities including those most vulnerable in the communities. People need access to WiFi and computers.
